```python
import numpy as np
import pandas as pd
from google.colab import files
import io
import zipfile
uploaded=files.upload()
zf=zipfile.ZipFile(io.BytesIO(next(iter(uploaded.values()))))
print(zf.namelist())
Batting_stats2016 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2016.csv'))
Batting_stats2017 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2017.csv'))
Batting_stats2018 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2018.csv'))
Batting_stats2019 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2019.csv'))
Batting_stats2020 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2020.csv'))
Batting_stats2021 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2021.csv'))
Batting_stats2022 = pd.read_csv(zf.open('IPL Player Stats/Batting Stats/BATTING STATS - IPL_2022.csv'))
Bowling_stats2016 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2016.csv'))
Bowling_stats2017 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2017.csv'))
Bowling_stats2018 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2018.csv'))
Bowling_stats2019 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2019.csv'))
Bowling_stats2020 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2020.csv'))
Bowling_stats2021 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2021.csv'))
Bowling_stats2022 = pd.read_csv(zf.open('IPL Player Stats/Bowling Stats/BOWLING STATS - IPL_2022.csv'))
Batting_stats2016.columns=Batting_stats2016.columns.str.strip().str.replace(" ", "_")
Batting_stats2017.columns=Batting_stats2017.columns.str.strip().str.replace(" ", "_")
Batting_stats2018.columns=Batting_stats2018.columns.str.strip().str.replace(" ", "_")
Batting_stats2019.columns=Batting_stats2019.columns.str.strip().str.replace(" ", "_")
Batting_stats2020.columns=Batting_stats2020.columns.str.strip().str.replace(" ", "_")
Batting_stats2021.columns=Batting_stats2021.columns.str.strip().str.replace(" ", "_")
Batting_stats2022.columns=Batting_stats2022.columns.str.strip().str.replace(" ", "_")
Bowling_stats2016.columns=Bowling_stats2016.columns.str.strip().str.replace(" ", "_")
Bowling_stats2017.columns=Bowling_stats2017.columns.str.strip().str.replace(" ", "_")
Bowling_stats2018.columns=Bowling_stats2018.columns.str.strip().str.replace(" ", "_")
Bowling_stats2019.columns=Bowling_stats2019.columns.str.strip().str.replace(" ", "_")
Bowling_stats2020.columns=Bowling_stats2020.columns.str.strip().str.replace(" ", "_")
Bowling_stats2021.columns=Bowling_stats2021.columns.str.strip().str.replace(" ", "_")
Bowling_stats2022.columns=Bowling_stats2022.columns.str.strip().str.replace(" ", "_")
print(Batting_stats2016.head())
print(Batting_stats2017.head())
print(Batting_stats2018.head())
print(Batting_stats2019.head())
print(Batting_stats2020.head())
print(Batting_stats2021.head())
print(Batting_stats2022.head())
print(Bowling_stats2016.head())
print(Bowling_stats2017.head())
print(Bowling_stats2018.head())
print(Bowling_stats2019.head())
print(Bowling_stats2020.head())
print(Bowling_stats2021.head())
print(Bowling_stats2022.head())
def check_dtypes_null(df):
  print(df.dtypes)
  print(df.isnull().sum())
check_dtypes_null(Batting_stats2016)
check_dtypes_null(Batting_stats2017)
check_dtypes_null(Batting_stats2018)
check_dtypes_null(Batting_stats2019)
check_dtypes_null(Batting_stats2020)
check_dtypes_null(Batting_stats2021)
check_dtypes_null(Batting_stats2022)
check_dtypes_null(Bowling_stats2016)
check_dtypes_null(Bowling_stats2017)
check_dtypes_null(Bowling_stats2018)
check_dtypes_null(Bowling_stats2019)
check_dtypes_null(Bowling_stats2020)
check_dtypes_null(Bowling_stats2021)
check_dtypes_null(Bowling_stats2022)
batting_datasets=[Batting_stats2016,Batting_stats2017,Batting_stats2018,Batting_stats2019,Batting_stats2020,Batting_stats2021,Batting_stats2022]
bowling_datasets=[Bowling_stats2016,Bowling_stats2017,Bowling_stats2018,Bowling_stats2019,Bowling_stats2020,Bowling_stats2021,Bowling_stats2022]
for df in batting_datasets:
  df['HS']=df ['HS'].str.replace('*', '').astype('int')
```

```python
import matplotlib.pyplot as plt
seasons=['2016', '2017', '2018', '2019', '2020', '2021', '2022']
total_runs=[
    Batting_stats2016['Runs'].sum(),
    Batting_stats2017['Runs'].sum(),
    Batting_stats2018['Runs'].sum(),
    Batting_stats2019['Runs'].sum(),
    Batting_stats2020['Runs'].sum(),
    Batting_stats2021['Runs'].sum(),
    Batting_stats2022['Runs'].sum(),
]
plt.figure(figsize=(8,8))
plt.pie(total_runs, labels=seasons, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Total runs per season", fontsize=16)
plt.show()
```

```python
famous_players=["Virat Kohli", "David Warner", "AB de Villiers",
                  "Rohit Sharma", "Shikhar Dhawan", "MS Dhoni"]
batting_combined = pd.concat(
    [Batting_stats2016, Batting_stats2017, Batting_stats2018,
     Batting_stats2019, Batting_stats2020, Batting_stats2021, Batting_stats2022]
)
batting_filtered=batting_combined[batting_combined['Player'].isin(famous_players)]
player_runs=batting_filtered.groupby('Player')['Runs'].sum().reset_index()
player_runs=player_runs.sort_values(by="Runs", ascending=False)
plt.figure(figsize=(10,6))
plt.bar(player_runs["Player"], player_runs["Runs"], color="lightgreen")
plt.title("Famous palyer runs", fontsize=16)
plt.xlabel("Player", fontsize=14)
plt.ylabel("Runs", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
#for wickets in every season
seasons=["2016", "2017", "2018", "2019", "2020", '2021', '2022']
total_Wkts=[
    Bowling_stats2016['Wkts'].sum(),
    Bowling_stats2017['Wkts'].sum(),
    Bowling_stats2018['Wkts'].sum(),
    Bowling_stats2019['Wkts'].sum(),
    Bowling_stats2020['Wkts'].sum(),
    Bowling_stats2021['Wkts'].sum(),
    Bowling_stats2022['Wkts'].sum(),
]
plt.figure(figsize=(10, 6))
plt.scatter(seasons, total_Wkts, color="orange", s=100)
plt.plot(seasons, total_Wkts, color="green", linestyle="-", label="Trend Line")#for adding line
plt.title("Total Wickets in Every Season", fontsize=16)
plt.xlabel("Seasons", fontsize=14)
plt.ylabel("Total Wickets", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
bowling_combined = pd.concat(
    [Bowling_stats2016, Bowling_stats2017, Bowling_stats2018,
     Bowling_stats2019, Bowling_stats2020, Bowling_stats2021, Bowling_stats2022]
)
top_players = bowling_combined.groupby('Player')['Wkts'].sum().sort_values(ascending=False).head(5).index  # Top 5 players
bowling_filtered = bowling_combined[bowling_combined['Player'].isin(top_players)]
plt.figure(figsize=(12, 8))
bowling_filtered.boxplot(column='Wkts', by='Player', grid=False, patch_artist=True,
                         boxprops=dict(facecolor='lightblue'))
plt.title('Wickets Distribution for Top 5 Players', fontsize=16)
plt.xlabel('Player', fontsize=14)
plt.ylabel('Wickets', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top_players = bowling_combined.groupby('Player')['Wkts'].sum().sort_values(ascending=False).head(5).index
batting_combined = pd.concat(
    [Batting_stats2016, Batting_stats2017, Batting_stats2018,
     Batting_stats2019, Batting_stats2020, Batting_stats2021, Batting_stats2022]
)
batting_filtered = batting_combined[batting_combined['Player'].isin(top_players)]
plt.figure(figsize=(12, 6))
sns.violinplot(x="Player", y="Mat", data=batting_filtered, palette="muted")
plt.title("Matches Distribution", fontsize=16)
plt.xlabel("Player", fontsize=14)
plt.ylabel("Matches", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
