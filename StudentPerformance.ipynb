```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google import colab
from google.colab import files
```

```python
uploaded=files.upload()
df=pd.read_csv('StudentsPerformance.csv')
df.isnull().sum()
df['test preparation course']=df['test preparation course'].replace('none', 'incomplete')
print(df.head())
df.info()
```

```python
#Gender distribution
gender_counts=df["gender"].value_counts()
avg=df.groupby('gender')[["math score", "reading score", "writing score"]].mean()
fig, axs=plt.subplots(1,2, figsize=(12,6))
axs[0].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=["pink", "lightblue"])
axs[0].set_title("Gender Distribution")
avg.plot(kind="bar", ax=axs[1])
axs[1].set_title("Average scores by Gender")
axs[1].set_ylabel("scores")
plt.tight_layout()
plt.show()
```

```python
#plot depicting divisions of students who has completed or not completed the course
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='test preparation course', palette='Set2')
plt.xlabel("test preparation course")
plt.ylabel('Number of students')
plt.show()
```

```python
#to compare score distributions by subject
plt.figure(figsize=(10,6))
plt.boxplot([df["math score"], df["reading score"], df["writing score"]], labels=["maths", "reading", "writing"], patch_artist=True)
plt.xlabel("Score distribution by subjects")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
#Parental level of eduction
plt.figure(figsize=(18,6))
sns.violinplot(data=df, x='parental level of education', y="math score", palette="Reds")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=45, ha='right')

sns.violinplot(data=df, x='parental level of education', y="reading score", palette="Greens")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=45, ha='right')

sns.violinplot(data=df, x='parental level of education', y="writing score", palette="Oranges")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=45, ha='right')
```

```python
# score comparison btw subjects
subject=df[["math score", 'reading score', "writing score"]].mean()
subject.plot(kind="line", marker="o", figsize=(10,6))
plt.title("score comparison btw subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```
