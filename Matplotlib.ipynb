**Line Plots**
Definition
Displays data points connected by a continuous line

Ideal for showing trends over time

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y, marker='o',
linestyle='-', color='b')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.show()
```

**Bar Charts**
Definition
Represents categorical data with rectangular bars

Bar length proportional to value

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color='g')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

**Histograms**
Definition
Represents distribution of numerical data

Divides data into bins and counts occurrences

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30,
color='purple', alpha=0.7)
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
#print(data)
#change random.randint(1,10,1000)
```

**Scatter Plots**
Definition
Displays values as points on Cartesian plane

Reveals relationships between variables

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='red',
marker='x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

**Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a sample dataset
np.random.seed(42)
df = pd.DataFrame(
    {
        "Category": np.random.choice(["A", "B", "C"], size=100),
        "Value1": np.random.normal(loc=50, scale=15, size=100),
        "Value2": np.random.normal(loc=30, scale=10, size=100),
        "Value3": np.random.normal(loc=70, scale=20, size=100),
    }
)


# Create plots
plt.figure(figsize=(6, 4))

# Box Plot
# plt.subplot(2, 2, 1)
sns.boxplot(x="Category", y="Value1", data=df)
plt.title("Box Plot")
plt.savefig(
    "sea_Boxplot.png",
    dpi=300,
    bbox_inches="tight",
    transparent=True,
)
```

```python
# Violin Plot
plt.figure(figsize=(6, 4))

sns.violinplot(x="Category", y="Value1", data=df)
plt.title("Violin Plot")
plt.savefig(
    "sea_violin.png",
    dpi=300,
    bbox_inches="tight",
    transparent=True,
)
```

```python
# Pair Plot
plt.figure(figsize=(6, 4))

pairplot = sns.pairplot(df, hue="Category")
plt.savefig(
    "sea_pair.png",
    dpi=300,
    bbox_inches="tight",
    transparent=True,
)
```

```python
plt.subplot(2, 2, 3)
# Select only numeric columns
numeric_df = df.select_dtypes(include=["float64", "int64"])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", center=0)
plt.title("Correlation Heatmap (Numeric Columns)")
plt.savefig(
    "sea_corr.png",
    dpi=300,
    bbox_inches="tight",
    transparent=True,
)
```
